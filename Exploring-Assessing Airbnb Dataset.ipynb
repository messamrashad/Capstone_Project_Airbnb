#### Assessing Phase

###### Quality Issues:
Based on the concepts of good quality data which are: Completeness, validity, accuracy, consistency. This dataset should be cleaned from the mentioned below issues.

For **Listing Dataframe**: </br>

1- Extract the needed columns only. </br>

2- Correct all listings with missing values in the extracted columns to the following Columns/Values: </br>

- **host_name**: "Not Provided" </br> 

- **host_since**: "Not Provided" </br> 

- **host_response_time**: "Not Provided" </br> 

- **host_is_superhost**: "Not Provided" </br> 

- **host_total_listings_count**: "0" </br> 

- **host_identity_verified**: "Not Provided" </br> 

- **review_scores_rating**: "0" </br> 

- **Cancellation_policy**: "No Cancellation Policy provided" </br> 

- **reviews_per_month**: "0" </br> 

- **host_identity_verified**: "f" </br> 

For **Reviews Dataframe**:

1- Correct all rows with Null values in Comments column to have "No Comment Provided". </br>



###### Tidiness Issues:

For **Calendar Dataframe**: </br>

1- Change the datatype of **Date** column to be "DataTime" column instead of "Object" column. </br>

2- Change the datatype of **Price, Adjusted_Price, Maximum_Nights, Minimum_Nights** column to be "DataTime" column instead of **"Object"** column. </br>

```python
calendar_original_df_clean.info()
```

```python
calendar_original_df_clean['date'] = pd.to_datetime(calendar_original_df_clean['date'], format= '%Y-%m-%d')
```

#### Convert The data types of Price, Adjusted_price, Maximum_nights, Minimum_Nights to be Integer.

```python
##Start to convert Price Column
calendar_original_df_clean['price'] = calendar_original_df_clean['price'].str.replace("$","")
calendar_original_df_clean['price'] = pd.to_numeric(calendar_original_df_clean['price'],errors='coerce')
calendar_original_df_clean['price'] = calendar_original_df_clean['price'].fillna(value=0)
calendar_original_df_clean['price'] = calendar_original_df_clean['price'].astype(np.int64)
```

```python
##Convert Adjusted_Price Column
calendar_original_df_clean['adjusted_price'] = calendar_original_df_clean['adjusted_price'].str.replace("$","")
calendar_original_df_clean['adjusted_price'] = pd.to_numeric(calendar_original_df_clean['adjusted_price'],errors='coerce')
calendar_original_df_clean['adjusted_price'] = calendar_original_df_clean['adjusted_price'].fillna(value=0)
calendar_original_df_clean['adjusted_price'] = calendar_original_df_clean['adjusted_price'].astype(np.int64)
```

```python
##Convert Maximum_Nights Column
calendar_original_df_clean['maximum_nights'] = pd.to_numeric(calendar_original_df_clean['maximum_nights'],errors='coerce')
calendar_original_df_clean['maximum_nights'] = calendar_original_df_clean['maximum_nights'].fillna(value=0)
calendar_original_df_clean['maximum_nights'] = calendar_original_df_clean['maximum_nights'].astype(np.int64)
```

```python
##Convert Minimum_Nights Column
calendar_original_df_clean['minimum_nights'] = pd.to_numeric(calendar_original_df_clean['minimum_nights'],errors='coerce')
calendar_original_df_clean['minimum_nights'] = calendar_original_df_clean['minimum_nights'].fillna(value=0)
calendar_original_df_clean['minimum_nights'] = calendar_original_df_clean['minimum_nights'].astype(np.int64)
```
